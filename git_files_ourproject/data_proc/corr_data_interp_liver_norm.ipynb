## Gastrocnemius Filling

```python
X_orig = np.loadtxt("/content/drive/MyDrive/ProjectX 2021/Code/Correlation Analysis/liver_mlratio_clean_nqrank.csv", delimiter=",")
```

```python
X = X_orig
```

```python
X.shape
```

```python
np.sort(np.where(np.isnan(X).any(axis=0))[0]).shape
```

```python
counter = 0
for i in np.sort(np.where(np.isnan(X).any(axis=0))[0]):
  rsq_vals = []
  gradients = []
  intercepts = []
  for j in [x for x in range(X.shape[1]) if x != i]:
    gradient, intercept, r_value, p_value, std_err = stats.linregress(X[~np.isnan(X[:, i]), i], X[~np.isnan(X[:, i]), j])
    rsq_vals.append(r_value**2)
    gradients.append(gradient)
    intercepts.append(intercept)
  rsq_vals = np.array(rsq_vals)
  gradients = np.array(gradients)
  intercepts = np.array(intercepts)
  print("\nBest Gene Match: ", i, np.nanargmax(rsq_vals), np.nanmax(rsq_vals), gradients[np.nanargmax(rsq_vals)], intercepts[np.nanargmax(rsq_vals)])
  nanargs = np.where(np.isnan(X[:, i]))
  print(nanargs[0])
  for k in nanargs[0]:
    counter += 1
    X[k, i] = gradients[np.nanargmax(rsq_vals)]*X[k, j] + intercepts[np.nanargmax(rsq_vals)]
    print(f"Missing Value {counter} filled!")
```
