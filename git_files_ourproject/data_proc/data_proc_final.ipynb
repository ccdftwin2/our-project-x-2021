## Glucose Interpolation

```python
df_gluc = pd.read_csv ('/content/drive/MyDrive/ProjectX 2021/Data sets for F2 mice/clinical traits on F2 mice/(F2)B6_BTBR OB mice_clinic_traits.csv')
```

```python
np.array(df_gluc[["Unnamed: 7", "Unnamed: 13", "Unnamed: 19", "Unnamed: 25"]][1:]).shape
```

```python
gluc_data = np.array(df_gluc[["Unnamed: 7", "Unnamed: 13", "Unnamed: 19", "Unnamed: 25"]][1:])
gluc_id= np.array(df_gluc[["Unnamed: 0"]][1:])
```

```python
gluc_id = gluc_id[~np.isnan(gluc_data[:, -1].astype(float))].astype(float)
```

```python
gluc_data = gluc_data[~np.isnan(gluc_data[:, -1].astype(float))].astype(float)
```

```python
for i in np.sort(np.where(np.isnan(gluc_data).any(axis=1))[0]):
  x = np.where(~np.isnan(gluc_data[i, :]))[0]
  xnew = np.where(np.isnan(gluc_data[i, :]))[0]
  y = gluc_data[i, :][x]
  gradient, intercept, r_value, p_value, std_err = stats.linregress(x, y)
  gluc_data[i, :][xnew] = gradient*xnew+intercept
  print(gluc_data[i, :], xnew)
```
