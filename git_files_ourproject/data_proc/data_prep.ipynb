## Adipose

```python
df_adi_mlratio['Unnamed: 0'] = [x[-4:] for x in df_adi_mlratio['Unnamed: 0']] 
df_adi_mlratio = df_adi_mlratio.rename(columns={"Unnamed: 0":"MouseID"})
```

```python
X = df_adi_mlratio.to_numpy(dtype=float)[:,1:] # Expression level: (num of mice, num of genes)
y = []                  # Glucose level: (num of mice, its glucose level from 4 wk - 10 wk)
sex = []                # Sex for each mice, 0 for male, 1 for female

for id in df_adi_mlratio['MouseID']:
  i = df_gluc.index[df_gluc['Unnamed: 0'] == id].tolist()
  if len(i)!= 0:
    i = i[0]
    y.append([df_gluc["Unnamed: 25"][i]])
    sex.append(0 if df_gluc["Unnamed: 1"][i] == "M" else 1)

sex = np.array(sex,dtype=int)
y = np.array(y,dtype=float)
y.shape, sex.shape, X.shape
```

```python
np.savetxt("/content/drive/MyDrive/ProjectX 2021/Code/Correlation Analysis/adi_mlratio_nqrank_clean_sex.csv", sex, delimiter=",")
```

## Combined Dataset

```python
df_adi_mlratio.columns = df_adi_mlratio.columns[:1].union("adi_" + df_adi_mlratio.columns[1:])
df_gas_mlratio.columns = df_gas_mlratio.columns[:1].union("gas_" + df_gas_mlratio.columns[1:])
df_islet_mlratio.columns = df_islet_mlratio.columns[:1].union("islet_" + df_islet_mlratio.columns[1:])
df_kid_mlratio.columns = df_kid_mlratio.columns[:1].union("kid_" + df_kid_mlratio.columns[1:])
df_liver_mlratio.columns = df_liver_mlratio.columns[:1].union("liver_" + df_liver_mlratio.columns[1:])
```

```python
m1 = pd.merge(df_adi_mlratio, df_gas_mlratio, on="MouseID")
m2 = pd.merge(m1, df_islet_mlratio, on="MouseID")
m3 = pd.merge(m2, df_kid_mlratio, on="MouseID")
m4 = pd.merge(m3, df_liver_mlratio, on="MouseID")
```

```python
X = m4.to_numpy(dtype=float)[:,1:] # Expression level: (num of mice, num of genes)
y = []                  # Glucose level: (num of mice, its glucose level from 4 wk - 10 wk)
sex = []                # Sex for each mice, 0 for male, 1 for female

for id in m4['MouseID']:
  i = df_gluc.index[df_gluc['Unnamed: 0'] == id].tolist()
  if len(i)!= 0:
    i = i[0]
    y.append([df_gluc["Unnamed: 25"][i]])
    sex.append(0 if df_gluc["Unnamed: 1"][i] == "M" else 1)

sex = np.array(sex,dtype=int)
y = np.array(y,dtype=float)
y.shape, sex.shape, X.shape
```
